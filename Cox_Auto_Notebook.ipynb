```python
import pandas as pd
from pandasql import sqldf
```

```python
# creating 3 data frames from each CSV file

df_Dealers = pd.read_csv('Dealers.csv')

df_Offers = pd.read_csv('Offers.csv')

df_Transactions = pd.read_csv('Transactions.csv')
```

```python
# Answer 1
query1 = "SELECT COUNT(DISTINCT OFFERVIN) FROM df_Offers WHERE OFFERDATE = '1/4/2022'"
df1 = sqldf(query1)
df1
```

```python
# Answer 2
query2 = "SELECT COUNT(DISTINCT OFFERVIN) FROM df_Offers WHERE REPLACEMENT_MAKE IS NOT NULL"
df2 = sqldf(query2)
df2
```

```python
# Answer 3
query3 = "SELECT * FROM df_Offers o LEFT JOIN df_Dealers d ON o.DEALER_ID = d.CUSTOMER_ID WHERE d.COMPANY_NAME = 'Team Gillman Subaru North'"
df3 = sqldf(query3)
df3
```

```python
# Answer 4
query4 = "SELECT DISTINCT * FROM df_Offers o LEFT JOIN df_Transactions t on o.OFFERVIN = t.OFFERVIN WHERE t.TRANSACTION_DATE >= '1/1/2022' AND t.TRANSACTION_DATE < '2/1/2022'"
df4 = sqldf(query4)
df4
```

```python
# Answer 5
# query5 = "SELECT VEHICLEMAKE, VEHICLEMODEL, SUM(OFFER_PRICE_FINAL) as Total_Offer_Price FROM df_Offers GROUP BY VEHICLEMAKE, VEHICLEMODEL ORDER BY VEHICLEMAKE, VEHICLEMODEL"
query5 = "WITH Make_Data as \
            ( \
            SELECT \
            VEHICLEMAKE, \
            SUM(OFFER_PRICE_FINAL) as Make_Total_Offer_Price \
            FROM df_Offers \
            GROUP BY VEHICLEMAKE\
            ) \
            , \
            Model_Data as \
            ( \
            SELECT \
            VEHICLEMAKE, \
            VEHICLEMODEL, \
            SUM(OFFER_PRICE_FINAL) as Total_Offer_Price \
            FROM df_Offers \
            GROUP BY \
            VEHICLEMAKE, \
            VEHICLEMODEL \
            ) \
            SELECT \
                mo.*, \
                (CAST(mo.Total_Offer_Price as float) / CAST(ma.Make_Total_Offer_Price as float) * 100) as Percentage_Of_Make_Total \
                FROM Make_Data ma \
                LEFT JOIN Model_Data mo on ma.VEHICLEMAKE = mo.VEHICLEMAKE \
                ORDER BY ma.VEHICLEMAKE, mo.VEHICLEMODEL"
df5 = sqldf(query5)
df5
```
